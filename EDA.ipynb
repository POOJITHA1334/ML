identifying outliers
to identify outliers the following methods can be used.
1. Z-score:z-score is also called standard core.this value/score helps to understand that how far is the data point from the mean.And after setting up the threshold value one can  utilize z-score values to define the outliers.
2. IQR(Inter Quartile Range):Inter Quateral range approach to finding the outliers is the most commonly used and most trusted approach used in research field.
3. x-x(mean)/standard deviation.

```python
#outlier detection using z-score
from scipy import stats #scipy is the library used for importing statistics methods
hp_zscore=np.abs(stats.zscore(data['normalized-losses']))
hp_zscore
```

```python
#set a threshold
threshold=hp_zscore.max()
```

```python
#position of the outlier
index_val=np.where(hp_zscore>=threshold)
data.iloc[index_val]['normalized-losses']
```

```python
#drop the row that has the outlier
new_data=data.drop(index=data.iloc[index_val]['normalized-losses'].index)
new_data
```

```python
#calculate the upper and lower limits
Q1=data['length'].quantile(0.25)
Q3=data['length'].quantile(0.75)
IQR = Q3-Q1
lower=Q1-1.5*IQR
upper=Q3+1.5*IQR
#create arrays of boolean values indicating the outlier rows
upper_arrays=np.where(data['length']>=upper)[0]
lower_arrays=np.where(data['length']<=lower)[0]
Q1
```

Grouping:
1. The 'groupby'method groups data by different categories.The data is grouped based on one or some variables and analysis is performed on the individual groups.
2. For example, let's groupby the variable "drive-wheels".we see that there are 3 different categories of wheels.



```python
df_group_one=data[['drive-wheels','price']]
df_group_one.head()
```

```python
#grouping results
df_gptest=data[['drive-wheels','price','body-style']]
grouped_test1=df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```
