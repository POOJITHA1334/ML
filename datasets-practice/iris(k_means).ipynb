```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
df=pd.read_csv("/content/Iris.csv")
df
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,[3,4]].values
x
```

```python
from sklearn.cluster import KMeans
b=[]
for i in range(1,10):
  kmeans=KMeans(n_clusters = i,init = 'k-means++',random_state=0)
  kmeans.fit(x)
  b.append(kmeans.inertia_)
```

```python
plt.plot(range(1,10),b)
plt.title("iris elbow graph")
plt.xlabel("PetalLengthCm")
plt.ylabel("PetalWidthCm")
plt.show()
```

```python
kmeansmodel = KMeans(n_clusters = 3, init='k-means++', random_state=0)
y_kmeans = kmeansmodel.fit_predict(x)
```

```python
plt.scatter(x[y_kmeans == 0,0],x[y_kmeans == 0,1],s=100,c='red',label='centroid')
plt.scatter(x[y_kmeans == 1,0],x[y_kmeans == 1,1],s=100,c='blue',label='centroid')
plt.scatter(x[y_kmeans == 2,0],x[y_kmeans == 2,1],s=100,c='green',label='centroid')
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1],marker='*',s=200,c='black',label='centroids')
plt.title('clusters of iris species')
plt.xlabel('petallength')
plt.ylabel('petalwidth')
plt.legend()
plt.show()
```
